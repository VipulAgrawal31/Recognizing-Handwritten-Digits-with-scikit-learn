```python
shuffle_index=np.random.permutation(6000)
x_train,y_train=x_train[shuffle_index],y_train[shuffle_index]
```

```python
y_train=y_train.astype(np.int8)
y_test=y_test.astype(np.int8)
y_train_8=(y_train==8)
y_test_8=(y_test==8)
```

```python
y_train_8
```

```python
y_test_8
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf= LogisticRegression(tol =0.1,solver='lbfgs')
```

```python
clf.fit(x_train,y_train_8)
```

```python
clf.predict([some_digit])
```

```python
from sklearn.model_selection import cross_val_score
a=cross_val_score(clf,x_train,y_train_8,cv=3,scoring='accuracy')
a
```

```python
a.mean()
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred=cross_val_predict(clf,x_train,y_train_8,cv=3)
```

```python
y_train_pred
```

```python
from sklearn.metrics  import confusion_matrix
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
y_scores=cross_val_predict(clf,x_train,y_train_8,cv=3,method='decision_function')
```

```python
y_scores
```

```python
precisions,recalls,thresholds=precision_recall_curve(y_train_8,y_scores)
```

```python
plt.plot(thresholds,precisions[:-1],"b--",label="Precision")
plt.plot(thresholds,recalls[:-1] ,"g-",label="Recall")
plt.xlabel("Thresholds")
plt.legend(loc="upper left")
plt.ylim([0,1])
plt.show()
```
